```python
# implementation de biblioth√®que 
import cv2
import os
import numpy as np
```

# √âtape 1 : Charger les iamges

```python
# acces a les images 
data_path = "dataSet/MVI_20012"

images = [f for f in os.listdir(data_path) if f.endswith(".jpg")]
images.sort()  # pour garde l oredre des frames 

print(f"{len(images)} images trouv√©es")
```

# Etape 2 : Pre-traitement (Pre-processing)

```python
preproc_path = "data_preproce"

os.makedirs(preproc_path,exist_ok = True)

# boucle sur chaque image
for i ,img_name in enumerate(images) :
    img_pt = os.path.join(data_path,img_name)

    # liree l image
    image = cv2.imread(img_pt)

    # redimensioner l'image 
    img_res = cv2.resize(image,(640,480) ) # largeur , hauteur

    # conversion en niveau de gris 
    gray = cv2.cvtColor(img_res,cv2.COLOR_BGR2GRAY)

    # appliquation de gaussien pour reduire le bruit 
    blur = cv2.GaussianBlur(gray,(5,5),0)
    # median reduire le bruit de sel 
    blur = cv2.medianBlur(gray,5)

    # normalization 
    normalized = blur /255.0
    # conv float to uint8 pour sauv
    save_img = (normalized * 255).astype(np.uint8)

    # Sauvegarde d image pre-traitee
    save_path = os.path.join(preproc_path, f"frame_{i:04d}.jpg")
    cv2.imwrite(save_path, save_img)

print("Pr√©-traitement termin√©, images sauvegard√©es")
```

```python
import torch
import cv2
import os
import warnings
warnings.filterwarnings('ignore') # This hides those messy deprecation warnings

# Load the model
model = torch.hub.load('ultralytics/yolov5', 'yolov5s', pretrained=True, trust_repo=True)

# les classes
vehicle_classes = ['car', 'bus', 'truck', 'motorbike']

# Dossier des images pr√©-trait√©es
preproc_path = "data_preproce"

images = sorted(os.listdir(preproc_path))

for img_name in images[:5]:  # tester sur 5 images
    img_path = os.path.join(preproc_path, img_name)

    # D√©tection
    results = model(img_path)

    # Affichage
    results.show()
    results.print()
```

# Charger YOLO et d√©tecter les v√©hicules

```python
total_vehicles = 0
images = sorted(os.listdir(preproc_path))[:100]

for i, img_name in enumerate(images):
    img_path = os.path.join(preproc_path, img_name)

    results = model(img_path, size=640)

    detections = results.pandas().xyxy[0]

    vehicles = detections[detections['name'].isin(vehicle_classes)]
    count = len(vehicles)

    print(f"[{i+1}/100] {img_name} ‚Üí {count} v√©hicules")

    total_vehicles += count

print("\nNombre total de v√©hicules d√©tect√©s :", total_vehicles)
```

# 3Ô∏è‚É£ Estimation de la congestion (OBJECTIF DU PROJET)

```python
def traffic_state(vehicle_count):
    if vehicle_count < 10:
        return "Fluide"
    elif vehicle_count < 20:
        return "Dense"
    else:
        return "Congestion"

for img_name in images:
    img_path = os.path.join(preproc_path, img_name)
    results = model(img_path)

    detections = results.pandas().xyxy[0]
    vehicles = detections[detections['name'].isin(vehicle_classes)]
    count = len(vehicles)

    state = traffic_state(count)
    print(f"{img_name} ‚Üí {count} v√©hicules ‚Üí Trafic : {state}")
```

# 4Ô∏è‚É£ Sauvegarder les r√©sultats (BONUS PROJET)

```python
save_result_path = "results"
os.makedirs(save_result_path, exist_ok=True)

for img_name in images[:10]:
    img_path = os.path.join(preproc_path, img_name)
    img = cv2.imread(img_path)

    results = model(img_path)
    detections = results.pandas().xyxy[0]
    vehicles = detections[detections['name'].isin(vehicle_classes)]

    count = len(vehicles)
    state = traffic_state(count)

    cv2.putText(img, f"Vehicles: {count} | {state}",
                (20,40), cv2.FONT_HERSHEY_SIMPLEX,
                1, (255,255,255), 2)

    cv2.imwrite(os.path.join(save_result_path, img_name), img)
```
