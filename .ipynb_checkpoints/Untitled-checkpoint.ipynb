# Étape 1 : Charger la vidéo MP4

```python
# implementation de bibliothèque 
import cv2
import os
```

```python
# Chemin de la vidéo
video_path = "dataSet/2.avi"

cap = cv2.VideoCapture(video_path)
os.makedirs("dataSet", exist_ok=True)

fps = int(cap.get(cv2.CAP_PROP_FPS))
step = fps

frame_id = 0
saved = 0
step = 10
max_frames = 1000

while saved < max_frames:
    ret, frame = cap.read()
    if not ret:
        break

    if frame_id % step == 0:
        frame = cv2.resize(frame, (640, 480))
        cv2.imwrite(f"dataSet/frame_{saved:04d}.jpg", frame)
        saved += 1

    frame_id += 1

cap.release()
print(f"{saved} frames sauvegardées")
```
